# Setup
First, we need to import the `json` and `requests` library so we can fetch and process JSON data from the RAC Collections API.

```python
import json
import requests
```

We'll also need to import some libraries that will help us to create the visualizations:

```python
%matplotlib inline

import matplotlib.pyplot as plt
import pandas as pd
import seaborn
```

# Getting Timeline Data

```python
def timeline_data(querystrings):
    years = []
    counts = []
    dataframe = pd.DataFrame({'year': list(range(1936, 2001))})
    for query in querystrings:
        query_counts = []
        for year in range(1936, 2001):
            hits = requests.get("http://10.0.1.90:8010/collections/2HnhFZfibK6SVVu86skz3k/minimap", params={"query": query, "start_date": year}).json()["hits"]
            query_counts.append(len(hits))
        dataframe[query] = query_counts
    return dataframe
```

```python
def draw_lineplots(data, **kwargs):
    seaborn.set_theme(palette="bright", font="monospace")
    ax = seaborn.lineplot(x='year', y='value', hue='variable', 
             data=pd.melt(data, ['year']))
    
draw_lineplots(timeline_data(["fellowship", "scholarship", "grants to individuals", "training", "exchange"]))
```

```python
def draw_lineplots(data, **kwargs):
    seaborn.set_theme(palette="bright", font="monospace")
    ax = seaborn.lineplot(x='year', y='value', hue='variable', 
             data=pd.melt(data, ['year']))
    
draw_lineplots(timeline_data(["university", "college", "association", "government of"]))
```

```python
def draw_lineplot(x, y, title, **kwargs):
    seaborn.set_theme(palette="bright", font="monospace")
    ax = seaborn.lineplot(
        x=x,
        y=y,
        **kwargs).set_title(title)
```

```python
draw_lineplot(*timeline_data("fellowship"))
plt.show()

draw_lineplot(*timeline_data("scholarship"))
plt.show()

draw_lineplot(*timeline_data("grants to individuals"))
plt.show()

draw_lineplot(*timeline_data("training"))
plt.show()

draw_lineplot(*timeline_data("exchange"))
plt.show()
```

```python
draw_lineplot(*timeline_data("university"))
plt.show()

draw_lineplot(*timeline_data("college"))
plt.show()

draw_lineplot(*timeline_data("association"))
plt.show()

draw_lineplot(*timeline_data("government of"))
plt.show()
```
